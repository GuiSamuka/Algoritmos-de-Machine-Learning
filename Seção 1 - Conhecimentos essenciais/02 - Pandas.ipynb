# Aula 03 - Criando DataFrame

```
data = {
    "Alunos": ['Caio', 'Jorge'],
    "Idade": [16, 15],
    "Nota": [8, 9]
}
```

```
df = pd.DataFrame(data, columns=['Alunos', 'Idade', 'Nota'])

display(df)
```

```
# Colher informações rápidas sobre o DataFrame

df.describe()
```

```
# Excluir uma linha

df = df.drop([0, 1]) # Remover a linha 0 e a 1
df
```

```
df = pd.DataFrame(data, columns=['Alunos', 'Idade', 'Nota'])
```

```
# Excluir uma coluna

df = df.drop('Idade', axis=1) # Remover a coluna 'Idade'
df
```

# Aula 05 - Dados estatísticos do DataFrame


```
data = {
    "Alunos": ['Caio', 'Jorge', 'Claudia'],
    "Idade": [16, 15, 16],
    "Nota": [8, 9, 7]
}
df = pd.DataFrame(data, columns=['Alunos', 'Idade', 'Nota'])
display(df)
```

```
# Maiores valores dentro do DataFrame
df['Idade'].max()

df.max()
```

```
# Menores valores dentro do DataFrame
df.min()
```

```
# Média dos valores das colunas
df.mean()
```

```
# Mediana do valores
df.median()
```

```
# Soma dos valores
df.sum() # Concatena os valores strings
```

# Aula 06 - Operações com DataFrame

```
data = {
    "Alunos": ['Caio', 'Jorge', 'Claudia'],
    "Idade": [16, 15, 16],
    "Nota": [8, 9, 7]
}
df = pd.DataFrame(data, columns=['Alunos', 'Idade', 'Nota'])
display(df)
```

### .add() -> Soma um determinado valor em todas as linhas de uma determinada coluna  

```
df.add(10)

# Não funciona com strings 
```

```
df['Nota'].add(10) # DataFrame não foi alterado
```

### .sub() -> Subtrai um determinado valor em todas as linhas de uma determinada coluna  

```
df['Idade'].sub(3)
```

### .mul() -> Multiplica um determinado valor em todas as linhas de uma determinada coluna  

```
df['Idade'].mul(10)
```

### .div() -> Multiplica um determinado valor em todas as linhas de uma determinada coluna  

```
df['Idade'].div(10)
```

### Acessar itens nas colunas

```
nota_um = df['Nota'][1]

print(nota_um)
```
