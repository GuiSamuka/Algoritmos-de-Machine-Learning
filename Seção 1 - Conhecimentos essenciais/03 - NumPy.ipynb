# Aula 09 - Arrays e Matrizes

```
import numpy as np # Manipulação de vetores e matrizes multidimensionais
```

### Vetores unidimensionais

```
x = np.array([1, 2, 3]) # Não é mutável, lembra a tupla (Lista são mutáveis)
print(x)
```

```
# Array é sempre de um tipo
y = np.array([1, 'caio', 3]) # Se um elemento for string, todos serão
print(y)
```

```
a = np.array([1, 2, 3])
b = np.array([1, 2, 3])

print(a + b) # Lista em python teria concatenado os dois vetores
```

```
x = np.zeros((5))
print(x)
```

```
x = np.ones((5))
print(x)
```

```
x = np.array([1, 2, 3, 4, 5, 6, 7])
print(x[2:])
```

### Vetores multidimensionais

```
x = np.array([[1, 2, 3], [1, 2, 3], [1, 2, 3]])
print(x)
```

```
x[0]
```

```
x[0][1]
```

```
x = np.array([[1, 2, 3]]*10)
print(x)
```

```
x = np.zeros((5, 5))
print(x)
```

```
x = np.ones((5, 5))
print(x)
```

```
# Para matriz quadratica (Matriz identidade)
x = np.eye(4)
print(x)
```

```
x = np.arange(5)
print(x)
```

```
x = np.arange(5, 10, 2)
print(x)
```
