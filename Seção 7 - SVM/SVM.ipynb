# 36 - Implementação do SVM

```
from sklearn import svm
import matplotlib.pyplot as plt
```

```
x = [[1, 2], [2, 1], [1, 1], [2, 2], [4, 4], [5, 4], [3, 5], [5, 6]]
y = [0, 0, 0, 0, 1, 1, 1, 1]
```

```
plt.scatter(x, y)
```

```
classificador = svm.SVC()
classificador.fit(x, y)
```

```
classificador.predict([[1, 3]])
```
