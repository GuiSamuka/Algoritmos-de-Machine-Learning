# Implementando Algoritmo de Regressão Linear

```
from sklearn.linear_model import LinearRegression
import numpy as np
```

```
idade = np.array([15, 16, 17, 18, 19, 20, 21, 22, 22, 28, 38, 45])
salario = np.array([600, 650, 800, 1000, 1200, 1200, 1350, 2300, 1500, 2300, 3000, 4000])

print(idade)
print(type(idade))
```

```
import matplotlib.pyplot as plt
plt.scatter(idade, salario)
plt.show()
```

```
idade = idade.reshape(-1, 1) # sklearn não aceita np.array, tem que mandar como Matriz
print(idade)
print(type(idade))
```

```
reg = LinearRegression()
reg = reg.fit(idade, salario)
```

Regressão Linear -> aproximar os dados por uma função de reta

$$ y = \omega_0 + \omega_1 \cdot x $$

```
reg.intercept_ # wo
```

```
reg.coef_ # w1
```

## Predições

### Forma manual

```
idade_Predict = 45

predict = reg.intercept_ + reg.coef_ * idade_Predict
predict  # esperado 4000
```

```
idade_Predict = 38

predict = reg.intercept_ + reg.coef_ * idade_Predict
predict  # esperado 3000
```

```
idade_Predict = 22

predict = reg.intercept_ + reg.coef_ * idade_Predict
predict  # esperado 2300 -> outlier, foge muito do gráfico
```

### Utilizando o sklearn

```
reg.predict([[45]])
```

# Visualizando o algoritmo

```
previsões = reg.predict(idade).reshape(-1, 1)  # Retorna as previsões para todas as idades
print(previsões, end='\n')
```

```
plt.scatter(idade, salario, color='red')
plt.plot(idade, previsões)

plt.show()
```
